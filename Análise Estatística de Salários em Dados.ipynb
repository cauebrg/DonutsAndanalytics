```python
# Etapa 01: Importação e Exploração inicial da base
import pandas as pd
```

```python
# Carreguei o arquivo CSV para o DataFame do pandas.
df = pd.read_csv('salario_profissionais_dados.csv')
```

```python
print("--- Primeiras 5 linhas do DataFrame ---")
print(df.head())
```

```python
# df.shape retorna uma tupla: (número de linhas, número de colunas).
print("\n--- Dimensões do DataFrame (linhas, colunas) ---")
print(df.shape)
```

```python
# Obtemos um resumo conciso do DataFrame, incluindo o tipo de dado de cada coluna
# e a contagem de valores não nulos.
print("\n--- Informações detalhadas do DataFrame ---")
df.info()
```

```python
# Verificamos a quantidade de valores ausentes (nulos) em cada coluna.

print("\n--- Contagem de valores nulos por coluna ---")
print(df.isnull().sum())

print("\n--- Análise Inicial Concluída! ---")
```

```python
# Etapa 02) Frequência e Distribuição das Categorias.
# Analisamos os cargos mais comuns na base de dados.
# 'value_counts()' conta a frequência de cada valor único na coluna 'job_title'.
# '.head(10)' mostra apenas os 10 cargos mais frequentes.

print("--- 10 Cargos mais comuns ---")
print(df['job_title'].value_counts().head(10))
```

```python
# Para ver a proporção (porcentagem) dos cargos, adicionamos 'normalize=True'.
print("\n--- Proporção dos 10 Cargos mais comuns (%) ---")
print(df['job_title'].value_counts(normalize=True).head(10) * 100)
```

```python
# Analisamos a distribuição dos níveis de experiência.
# O documento 'Orientação.pdf' nos dá a legenda para esses códigos:
# EN: Entry-level / Iniciante
# MI: Mid-level / Intermediário
# SE: Senior-level / Sênior
# EX: Executive-level / Executivo
print("\n--- Distribuição dos Níveis de Experiência ---")
print(df['experience_level'].value_counts())
print("\n--- Proporção dos Níveis de Experiência (%) ---")
print(df['experience_level'].value_counts(normalize=True) * 100)
```

```python
# Analisamos a distribuição do tamanho da empresa.
# O documento 'Orientação.pdf' nos dá a legenda para esses códigos:
# S: Pequena empresa
# M: Média empresa
# L: Grande empresa
print("\n--- Distribuição do Tamanho da Empresa ---")
print(df['company_size'].value_counts())
print("\n--- Proporção do Tamanho da Empresa (%) ---")
print(df['company_size'].value_counts(normalize=True) * 100)
```

```python
print("\n--- Análise de Frequências Concluída! ---")
```

```python
# Etapa 03) Estatísticas Descritivas de Salários.

# Aprofundamos na análise da variável 'salary_in_usd'.
# Calculamos as principais medidas de tendência central e dispersão.
print("--- Estatísticas Descritivas do Salário Anual em USD ---")
print(df['salary_in_usd'].describe())
```

```python
# Para visualizar a distribuição geral dos salários, criamos um histograma.
# Precisamos importar a biblioteca matplotlib para criar gráficos.
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6)) # Define o tamanho do gráfico para melhor visualização.
plt.hist(df['salary_in_usd'], bins=50, edgecolor='black') # 'bins' define o número de "barras" no histograma.
plt.title('Distribuição dos Salários Anuais em USD') # Título do gráfico.
plt.xlabel('Salário em USD') # Rótulo do eixo X.
plt.ylabel('Frequência') # Rótulo do eixo Y.
plt.grid(axis='y', alpha=0.75) # Adiciona uma grade no eixo Y.
plt.show() # Exibe o gráfico.
```

```python
# Comparações de salários por nível de experiência.
# Usamos 'groupby()' para agrupar os dados por 'experience_level' e depois calculamos a média do salário. [cite: 236]
print("\n--- Média Salarial por Nível de Experiência ---")
print(df.groupby('experience_level')['salary_in_usd'].mean())
```

```python
# Para uma visualização mais clara das comparações, usamos um boxplot. [cite: 234]
# Boxplots são excelentes para mostrar a distribuição (mediana, quartis, outliers) por grupo.
plt.figure(figsize=(12, 7))
df.boxplot(column='salary_in_usd', by='experience_level', grid=False)
plt.title('Distribuição de Salários por Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário em USD')
plt.suptitle('') # Remove o título automático do boxplot que duplica o nosso.
plt.show()
```

```python
print("\n--- Estatísticas Descritivas de Salários Concluída! ---")
```

```python
# Etapa 04) Comparações por País

# Verificamos os 10 países com as maiores médias salariais. [cite: 242]
# Agrupamos por 'company_location' e calculamos a média salarial.
# Em seguida, ordenamos os resultados em ordem decrescente ('ascending=False')
# para ver os maiores primeiro e pegamos os 10 primeiros com '.head(10)'. [cite: 245]
print("--- 10 Países com as Maiores Médias Salariais (Localização da Empresa) ---")
top_10_countries = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
print(top_10_countries)
```

```python
# Podemos também visualizar isso com um gráfico de barras para facilitar a comparação.
plt.figure(figsize=(14, 7))
top_10_countries.plot(kind='bar')
plt.title('Top 10 Países com Maiores Médias Salariais (Localização da Empresa)')
plt.xlabel('País')
plt.ylabel('Média Salarial em USD')
plt.xticks(rotation=45, ha='right') # Rotaciona os rótulos do eixo X para melhor leitura.
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Ajusta o layout para evitar sobreposição.
plt.show()
```

```python
print("\n--- Comparações por País Concluída! ---")
```

```python
# Etapa 05) Correlações e Tendências.

# Selecionamos as variáveis numéricas que queremos analisar a correlação.
# No nosso caso: 'salary_in_usd', 'work_year' e 'years_of_experience'.
numeric_cols = ['salary_in_usd', 'work_year', 'years_of_experience']
```

```python
# Calculamos a matriz de correlação entre essas colunas.
# A função '.corr()' calcula o coeficiente de correlação de Pearson por padrão,
# que mede a força e direção de uma relação linear.
correlation_matrix = df[numeric_cols].corr()
```

```python
print("--- Matriz de Correlação ---")
print(correlation_matrix)
```

```python
# Para uma visualização mais intuitiva da correlação, podemos usar um mapa de calor.
# Precisamos importar a biblioteca seaborn, que é ótima para visualização estatística.
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f") # 'annot=True' mostra os valores no mapa.
plt.title('Matriz de Correlação entre Salário, Ano e Experiência')
plt.show()
```

```python
# Interpretando os coeficientes: [cite: 252]
print("\n--- Interpretação dos Coeficientes de Correlação ---")
# Coeficiente próximo de 1: Correlação positiva forte (ex: uma aumenta, a outra também).
# Coeficiente próximo de -1: Correlação negativa forte (ex: uma aumenta, a outra diminui).
# Coeficiente próximo de 0: Ausência de correlação linear (não significa que não há relação, apenas não linear).
```

```python
print(f"Correlação entre Salário (salary_in_usd) e Ano de Trabalho (work_year): {correlation_matrix.loc['salary_in_usd', 'work_year']:.2f}")
print("Isso indica se há uma tendência geral de aumento salarial ao longo dos anos registrados na base.")
print("\nObservação: Como o 'work_year' pode estar mais relacionado ao ano do registro do dado do que a uma tendência salarial ao longo do tempo para o *mesmo* profissional, é importante ter cautela na interpretação direta como 'aumento salarial ao longo dos anos para o mesmo cargo'.")
```

```python
print(f"\nCorrelação entre Salário (salary_in_usd) e Anos de Experiência (years_of_experience): {correlation_matrix.loc['salary_in_usd', 'years_of_experience']:.2f}")
print("Isso indica se, na nossa base de dados, profissionais com mais anos de experiência tendem a ter salários mais altos.")
```

```python
print("\n--- Análise de Correlações e Tendências Concluída! ---")
```

# Análise de Salários de Profissionais de Dados

## Etapa 01: Importação e Exploração Inicial da Base

```python
# Importamos a biblioteca pandas, essencial para trabalhar com dados tabulares.
# 'pd' é um apelido comum para pandas, facilitando a escrita do código.
import pandas as pd

# Carregamos o arquivo CSV para um DataFrame do pandas.
# Certifique-se de que o nome do arquivo 'salario_profissionais_dados.csv'
# esteja correto e que o upload foi feito no Colab.
df = pd.read_csv('salario_profissionais_dados.csv')

# Exibimos as 5 primeiras linhas do DataFrame.
print("--- Primeiras 5 linhas do DataFrame ---")
print(df.head())

# Verificamos o número de linhas e colunas do DataFrame.
print("\n--- Dimensões do DataFrame (linhas, colunas) ---")
print(df.shape)

# Obtemos um resumo conciso do DataFrame, incluindo o tipo de dado de cada coluna
# e a contagem de valores não nulos.
print("\n--- Informações detalhadas do DataFrame ---")
df.info()

# Verificamos a quantidade de valores ausentes (nulos) em cada coluna.
print("\n--- Contagem de valores nulos por coluna ---")
print(df.isnull().sum())

print("\n--- Análise Inicial Concluída! ---")
```

### Interpretação da Etapa 01:

Após a importação, verificamos que o dataset contém `[número de linhas]` linhas e `[número de colunas]` colunas. As colunas `work_year`, `salary_in_usd` e `years_of_experience` são numéricas, enquanto as demais são categóricas. ## Etapa 02: Frequência e Distribuição das Categorias

## Etapa 03: Estatísticas Descritivas de Salários

```python
# Aprofundamos na análise da variável 'salary_in_usd'.
print("--- Estatísticas Descritivas do Salário Anual em USD ---")
print(df['salary_in_usd'].describe())

# Para visualizar a distribuição geral dos salários, criamos um histograma.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['salary_in_usd'], bins=50, edgecolor='black')
plt.title('Distribuição dos Salários Anuais em USD')
plt.xlabel('Salário em USD')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Comparações de salários por nível de experiência.
print("\n--- Média Salarial por Nível de Experiência ---")
print(df.groupby('experience_level')['salary_in_usd'].mean())

# Para uma visualização mais clara das comparações, usamos um boxplot.
plt.figure(figsize=(12, 7))
df.boxplot(column='salary_in_usd', by='experience_level', grid=False)
plt.title('Distribuição de Salários por Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário em USD')
plt.suptitle('')
plt.show()

print("\n--- Estatísticas Descritivas de Salários Concluída! ---")
```

## Etapa 04: Comparações por País

```python
# Verificamos os 10 países com as maiores médias salariais.
print("--- 10 Países com as Maiores Médias Salariais (Localização da Empresa) ---")
top_10_countries = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
print(top_10_countries)

# Podemos também visualizar isso com um gráfico de barras para facilitar a comparação.
import matplotlib.pyplot as plt # Importa novamente para garantir que está disponível
plt.figure(figsize=(14, 7))
top_10_countries.plot(kind='bar')
plt.title('Top 10 Países com Maiores Médias Salariais (Localização da Empresa)')
plt.xlabel('País')
plt.ylabel('Média Salarial em USD')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

print("\n--- Comparações por País Concluída! ---")
```

## Etapa 05: Correlações e Tendências

```python
# Selecionamos as variáveis numéricas que queremos analisar a correlação.
numeric_cols = ['salary_in_usd', 'work_year', 'years_of_experience']

# Calculamos a matriz de correlação entre essas colunas.
correlation_matrix = df[numeric_cols].corr()

print("--- Matriz de Correlação ---")
print(correlation_matrix)

# Para uma visualização mais intuitiva da correlação, podemos usar um mapa de calor.
import seaborn as sns # Importa novamente para garantir que está disponível
import matplotlib.pyplot as plt # Importa novamente para garantir que está disponível

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Salário, Ano e Experiência')
plt.show()

# Interpretando os coeficientes:
print("\n--- Interpretação dos Coeficientes de Correlação ---")

print(f"Correlação entre Salário (salary_in_usd) e Ano de Trabalho (work_year): {correlation_matrix.loc['salary_in_usd', 'work_year']:.2f}")
print("Isso indica se há uma tendência geral de aumento salarial ao longo dos anos registrados na base.")

print(f"\nCorrelação entre Salário (salary_in_usd) e Anos de Experiência (years_of_experience): {correlation_matrix.loc['salary_in_usd', 'years_of_experience']:.2f}")
print("Isso indica se, na nossa base de dados, profissionais com mais anos de experiência tendem a ter salários mais altos.")

print("\n--- Análise de Correlações e Tendências Concluída! ---")
```
